
#1-Generate a sample by equation


> *y=w1X1+w2X2+....+b*





#2-Generate random number for Xs

#3-Apply Linear Regression

#4-Proof the same equatio will return using generated sample datasets



```python
import numpy as np
import pandas as pd

b= pd.DataFrame(np.ones((100,1)))
```

#Apply that Algorithm for 1 D , 2 D hyper plain


```python
#adding noise to randomize y
noise=pd.DataFrame(np.random.uniform(0,1,100))

#2D
x_2D= pd.DataFrame(np.random.randint(10, size=(100,2)))
x_2D.insert(0,"Ones",np.ones((100,1),dtype=int),True) #to make it augmented

#assuming w1=2 and w2=3 and b=1
y_actual=pd.DataFrame(2*x_2D[0]+3*x_2D[1]+b[0]) #y actual, training sample
y_test=noise+y_actual
# w=(x_2D^T * x_2D)^-1 * x_2D^T * y
w_estimated= pd.DataFrame(np.dot(np.dot(np.linalg.inv(np.dot(x_2D.transpose(), x_2D)), x_2D.transpose()), y_test))

print('\nthe model is proved to be accurate as w1 and w2 are approx_2Dimately equal to the assumed ones of 2,3 respectively')
w_estimated
```

```python
#1D
x_1D= pd.DataFrame(np.random.randint(10, size=(100,1)))
x_1D.insert(0,"Ones",np.ones((100,1),dtype=int),True) #to make it augmented

#assuming w1=5 and b=1
y_actual_1D=pd.DataFrame(5*x_1D[0]+b[0]) #y actual, training sample
y_test_1D=noise+y_actual_1D
# w=(x_1D^T * x_1D)^-1 * x_1D^T * y
w_estimated_1D= pd.DataFrame(np.dot(np.dot(np.linalg.inv(np.dot(x_1D.transpose(), x_1D)), x_1D.transpose()), y_test_1D))

print('\nthe model is proved to be accurate as w1 is approximately equal to the assumed one: 5')
w_estimated_1D
```
